```python
import matplotlib.pyplot as plt
import os
%matplotlib inline
import random
import shutil
```

```python
from IPython.display import clear_output
```

```python
def save_images(images_list,source_dir,dest_dir):
    if os.path.exists(dest_dir)==False:
        os.mkdir(dest_dir)
    for image in images_list:
        shutil.copy(os.path.join(source_dir,image), dest_dir)
```

```python
image_dir = 'C:\\Users\\sadok.guermazi\\Documents\\Projects\\openfoodfacts-ai\\nutrition-table-extraction\\notebooks\\data\\image_files\\'
output_folder = '../data/'
```

```python
images = [x for x in os.listdir(image_dir) if x.split('.')[-1]=='jpg']
```

```python
file = open("files.txt","w") 
file.write('\n'.join([x for x in os.listdir(image_dir) if x.split('.')[-1]=='jpg']) )
file.close() 
```

```python
random.shuffle(images)
```

```python
selected_images = []
```

```python
try:    
    while True:
            i=0
            for image in images:
                print("number of images selected : {}".format(i))
                plt.ion()
                plt.figure(figsize=(20,10))
                plt.imshow(plt.imread(os.path.join(image_dir,image)))
                plt.show()
                user_input = input('Keep ? (y if yes, enter if no)')
                if user_input.lower()=='y':
                    i+=1
                    selected_images.append(image)
                clear_output(wait=False)
except KeyboardInterrupt:
    pass
```

```python
save_images(selected_images,image_dir,output_folder)
```
